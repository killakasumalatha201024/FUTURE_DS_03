# **üîç Exploratory Analysis (Ratings Distribution)**

```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Sample DataFrame
data = {
    'Feedback': [
        "The event was fantastic!",
        "I didn't like the organization.",
        "",
        None,
        "It was okay, nothing special."
    ]
}

df = pd.DataFrame(data)

# Initialize VADER analyzer
analyzer = SentimentIntensityAnalyzer()

# Function to get sentiment
def get_sentiment(text):
    if pd.isnull(text) or text.strip() == "":
        return "Neutral"
    score = analyzer.polarity_scores(text)['compound']
    if score > 0.05:
        return "Positive"
    elif score < -0.05:
        return "Negative"
    else:
        return "Neutral"

# Apply sentiment analysis
df['Sentiment'] = df['Feedback'].apply(get_sentiment)

# Display results
print("Sentiment distribution:")
print(df['Sentiment'].value_counts())

print("\nData with sentiment:")
print(df)
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Pie chart of sentiment share
plt.figure(figsize=(6,4))
df['Sentiment'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    colors=["lightgreen", "salmon", "lightgrey"]
)
plt.title("Sentiment Share of Feedback")
plt.ylabel("")  # removes the y-label
plt.show()

# Word cloud from feedback (exclude empty/null)
text = " ".join(feedback for feedback in df['Feedback'].dropna() if feedback.strip() != "")
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(text)

plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Most Common Feedback Words")
plt.show()
```
